```python
from pyspark.sql import SparkSession
spark = SparkSession.builder \
        .appName("PracticePySpark") \
        .getOrCreate()
```

```python
df = spark.read.csv('Salaries.csv', header=True, inferSchema=True)
df = df.drop('Notes', 'Status')
df.show()
```

## 5. Tên của người được trả cao nhất (bao gồm benefits) là gì?


```python
# YOUR CODE HERE
from pyspark.sql.functions import max
maxPaid = df.agg(max('TotalPayBenefits')).collect()[0][0]
df.filter(df['TotalPayBenefits'] == maxPaid).select('EmployeeName').show()
```

## 6. Tên vị trí công việc (Job Title) của người đó là gì?


```python
# YOUR CODE HERE
df.filter(df['TotalPayBenefits'] == maxPaid).select('JobTitle').show()
```

## 7. Tên của người được trả thấp nhất (bao gồm benefits) là gì?

```python
# YOUR CODE HERE
from pyspark.sql.functions import min
minPaid = df.agg(min('TotalPayBenefits')).collect()[0][0]
df.filter(df['TotalPayBenefits'] == minPaid).select('EmployeeName').show()
```

## 8. Lương của người được trả thấp nhất (bao gồm benefits) là bao nhiêu?
**In cả dòng dữ liệu về người đó**

```python
# YOUR CODE HERE
df.filter(df.TotalPayBenefits == minPaid).show()
#print(minPaid)
```

## 9. In tất cả các dòng của các nhân viên có cùng Job Title với Joe Lopez.

```python
# YOUR CODE HERE
lopezRow = df.filter(df['EmployeeName'] == 'Joe Lopez').first()
lopezJobTitle = lopezRow['JobTitle']
df.filter(df['JobTitle'] == lopezJobTitle).show()
```

## 10. Lọc ra tất cả những người có TotalPay cao hơn TotalPay trung bình.

```python
# YOUR CODE HERE
avgTotalPay = df.agg({'TotalPay':'mean'}).collect()[0][0]
df.filter(df['TotalPay'] > avgTotalPay).show()
```
